```python
# extracting all the text present in the resume

import PyPDF2
def extract_text_from_pdf(file_path):
  with open(file_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ''
    for page_num in range(len(pdf_reader.pages)):
      page = pdf_reader.pages[page_num]
      text += page.extract_text()
  return text
```

```python
import json
import spacy
```

```python
# using Name Entitiy Recognition to classify the extracted text

class EntityGenerator(object):

    _slots__ = ['text']

    def __init__(self, text=None):
        self.text = text

    def get(self):
        """
        Return a Json
        """
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(self.text)
        text = [ent.text for ent in doc.ents]
        entity = [ent.label_ for ent in doc.ents]

        from collections import Counter
        import json

        data = Counter(zip(entity))
        unique_entity = list(data.keys())
        unique_entity = [x[0] for x in unique_entity]

        d = {}
        for val in unique_entity:
            d[val] = []

        for key,val in dict(zip(text, entity)).items():
            if val in unique_entity:
                d[val].append(key)
        return d
```

```python
corpus = skills1 + [text.split() for text in namess]
```

```python
model = Word2Vec(sentences = corpus, vector_size=50, window = 2, min_count=1, sg=1)
```

```python
# Function to vectorize text using Word2Vec model
def vectorize_text(text, model):
    vector = np.zeros(model.vector_size)
    count = 0
    for word in text:
        if word in model.wv:
            vector += model.wv[word]
            count += 1
    if count != 0:
        vector /= count
    return vector
```

```python
#vectorize descriptions and skills
vectorized_skills1 = vectorize_text(skills1, model)
description_vectors = [vectorize_text(text.split(), model) for text in namess]
```

```python
similarity_scores = [cosine_similarity([vectorized_skills1], [vector])[0][0] for vector in description_vectors]
```

#Custom Function

```python
#extracting keywords from the pdf

def keywords(tex):
    help = EntityGenerator(text=tex)
    responsees = help.get()

    # Iterate over ORG entities
    if 'ORG' in responsees:
        org_entities = responsees['ORG']
        for i in range(len(org_entities)):
            # Check if 'AI' is in the entity
            if 'AI' in org_entities[i]:
                # Append 'Artificial Intelligence' and 'Machine Learning'
                org_entities[i].replace('AI','')
                org_entities += ['Artificial Intelligence', 'Machine Learning']
            # Replace 'ML' with 'Machine Learning' and 'Deep Learning'
            if 'ML' in org_entities[i]:
                org_entities[i] = org_entities[i].replace('ML', '')
                org_entities+=['Machine Learning', 'Deep Learning']
            # Remove 'Engineering'
            org_entities[i] = org_entities[i].replace('Engineering', '')

    return responsees
```

```python
#using the keywords for recommending tests

def final(resume_path):

  # dict_output = json.loads(data_extract(resume_path))
  data_n = extract_words(keywords(extract_text_from_pdf(resume_path)))
  # data_n = ' '.join(words)

  test_names = tests['Test Name']

  corpus_n = data_n + [text.split() for text in test_names]
  model_n = Word2Vec(sentences = corpus_n, vector_size=50, window = 2, min_count=1, sg=1)

  vectorized_skills = vectorize_text(data_n, model_n)
  description_vectors_n = [vectorize_text(text.split(), model_n) for text in test_names]

  similarity_scores_n = [cosine_similarity([vectorized_skills], [vector])[0][0] for vector in description_vectors_n]
  tests['similarity_scores']=similarity_scores_n
  sorted_tests_n = tests.sort_values(by = 'similarity_scores', ascending = False)

  return sorted_tests_n.head(10)

```
